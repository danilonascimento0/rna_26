### Analise de sentimento do dataset IMDB utilizando Hugging Face.

```python
from datasets import load_dataset

dataset = load_dataset("stanfordnlp/imdb")
print(dataset)
```

### Instanciando o pipeline de análise de sentimento com o modelo `distilbert-base-uncased-finetuned-sst-2-english` da Hugging Face.

```python
from transformers import pipeline

classifier = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english",
    truncation=True
)
```

### Testando o pipeline com algumas frases de exemplo.

```python
results = classifier(["This is a great movie", "This is a bad movie"])
for result in results:
  print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
```

### Calculando a acurácia do modelo no conjunto de teste do dataset IMDB.

```python
test_data = dataset["test"]
labels = test_data["label"]

texts = [str(t) for t in test_data["text"] if t is not None]

results = classifier(texts, batch_size=32, truncation=True)

preds = [1 if r["label"] == "POSITIVE" else 0 for r in results]

accuracy = sum(p == y for p, y in zip(preds, labels)) / len(labels)

print("Accuracy:", accuracy)
```
